```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df=sns.load_dataset('titanic')
```

```python
df.head()
```

```python
df.columns
```

```python
df.info()
```

```python
df.drop(["deck","embark_town","alive","class","who","adult_male"],axis=1,inplace=True)
```

```python
df.info()
```

```python
df["age"].fillna(df["age"].mean(),inplace=True)
```

```python
df.dropna(subset=["embarked"],inplace=True)
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
df.head()
```

```python
df["sex"]
```

```python
df["sex"]=le.fit_transform(df["sex"])
```

```python
df["sex"]
```

```python
df["embarked"]=le.fit_transform(df["embarked"])
```

```python
df.head()
```

```python
df=df.astype(int)
```

```python
df.head()
```

```python
X=df.drop("survived",axis=1)
Y=df["survived"]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(X_train,Y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
accuracy_score(Y_test,y_pred)
```

```python
confusion_matrix(Y_test,y_pred)
```

```python
print(classification_report(Y_test,y_pred))
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
```

```python
X_train_scaled=sc.fit_transform(X_train)
X_test_scaled=sc.transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn_model=KNeighborsClassifier(n_neighbors=5)
knn_model.fit(X_train_scaled,Y_train)
```

```python
y_knn_pred=knn_model.predict(X_test_scaled)
```

```python
accuracy_score(Y_test,y_knn_pred)
```

```python
confusion_matrix(Y_test,y_knn_pred)
```

```python
classification_report(Y_test,y_knn_pred)
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
modle_NB=GaussianNB()
modle_NB.fit(X_train,Y_train)
```

```python
y_pred_nb=modle_NB.predict(X_test)
y_pred_nb
```

```python
accuracy_score(Y_test,y_pred_nb)
```

```python
confusion_matrix(Y_test,y_pred_nb)
```

```python
classification_report(Y_test,y_pred_nb)
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc=DecisionTreeClassifier()
dtc.fit(X_train_scaled,Y_train)
```

```python
Y_pred_dtc=dtc.predict(X_test_scaled)
```

```python
Y_pred_dtc
```

```python
accuracy_score(Y_test,Y_pred_dtc)
```

```python
confusion_matrix(Y_test,Y_pred_dtc)
```

```python
classification_report(Y_test,Y_pred_dtc)
```

```python
from sklearn.svm import SVC
model_svm=SVC(kernel='rbf')
model_svm.fit(X_train_scaled,Y_train)
```

```python
y_pred_svc=model_svm.predict(X_test_scaled)
```

```python
accuracy_score(Y_test,y_pred_svc)
```

```python
confusion_matrix(Y_test,y_pred_svc)
```

```python
classification_report(Y_test,y_pred_svc)
```
